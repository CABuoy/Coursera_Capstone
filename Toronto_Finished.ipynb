```python
#import library to open URL
import urllib.request
```

```python
#specify wikipedia url
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
```

```python
#put html into variable
page = urllib.request.urlopen(url)
```

```python
#import BeautifulSoup library
from bs4 import BeautifulSoup
```

```python
#parse HTML into specified format
soup = BeautifulSoup(page, 'lxml')
```

```python
#explore html data to find desired table
print(soup.prettify())
```

```python
#explore tables in data
tables = soup.find_all('table')
tables
```

```python
#select the table
the_table = soup.find('table', class_ = 'wikitable sortable')
the_table
```

```python
#cycle through rows of table and pick out cell values and assign to lists
A = []
B = []
C = []

for row in the_table.find_all('tr'):
    cell = row.find_all('td')
    if len(cell) == 3:
        A.append(cell[0].find(text=True)[0:-1]) # [0:-1] to remove /n from text
        B.append(cell[1].find(text=True)[0:-1]) # [0:-1] to remove /n from text
        C.append(cell[2].find(text=True)[0:-1]) # [0:-1] to remove /n from text
```

```python
# make list of column names
D = [v.find(text=True)[0:-1] for v in the_table.find_all('tr')[0].find_all('th')]
D
```

```python
#create pandas dataframe
import pandas as pd
df = pd.DataFrame(columns = D)
```

```python
#set column values equal to lists
df['Postal Code'] = A
df['Borough'] = B
df['Neighborhood'] = C
df
```

```python
#strip all data in dataframe (excludes headers)
df_trimmed = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
#strip column names
df_trimmed.columns = df_trimmed.columns.str.strip()
```

```python
df_trimmed['Borough'].value_counts() #check 'Not assigned' is only null value representation
```

```python
#drop all rows with 'Not assigned' value in Borough per instructions
#make values lower to account
df_trimmed.drop(df_trimmed.index[df_trimmed['Borough'] == 'Not assigned'],
                inplace = True)
```

```python
#reset index of cleaned up dataframe
df_trimmed.reset_index(drop = True, inplace = True)
df_trimmed
```

```python
#boolean list of any 'not assigned' neighborhood values
n_neigh = df_trimmed['Neighborhood'] == 'Not assigned'
```

```python
#set any 'not assigned' neighborhood values to corresponding borough value
df_trimmed['Neighborhood'][n_neigh] = df_trimmed['Borough'][n_neigh]
```

```python
#define function to retrieve dataframe of food venues in a neighborhood
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            100)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])
    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Postal Code', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
toronto_venues = getNearbyVenues(toronto_df['Postal Code'], toronto_df['Latitude'], toronto_df['Longitude'])
```

## Do some one-hot encoding

```python
# one hot encoding
toronto_onehot = pd.get_dummies(toronto_venues[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
toronto_onehot['Postal Code'] = toronto_venues['Postal Code'] 

# move neighborhood column to the first column
fixed_columns = [toronto_onehot.columns[-1]] + list(toronto_onehot.columns[:-1])
toronto_onehot = toronto_onehot[fixed_columns]

toronto_onehot.head()
```

## Now convert one-hot encoding info into a table with a row for each postal code, and each venue type as an attribute dictation the ratio of that venue type to all venues in postal code

```python
toronto_grouped = toronto_onehot.groupby('Postal Code').mean().reset_index()
toronto_grouped
```

# Cluster Neighborhoods based on venue category ratios

```python
# Matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors

# import k-means from clustering stage
from sklearn.cluster import KMeans
# set number of clusters
kclusters = 10

toronto_grouped_clustering = toronto_grouped.drop('Postal Code', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_
```

```python
#add kmeans categories to data frame
toronto_df['K-Means Category'] = kmeans.labels_
```

```python
toronto_df
```

## Create map of clusters

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(toronto_df['Latitude'], toronto_df['Longitude'], toronto_df['Postal Code'], toronto_df['K-Means Category']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
